```python
import requests
import csv
import time
from requests.exceptions import RequestException


def check_url_for_keyword(url, keyword, headers, timeout=10):
    """Check a single URL. Returns 1, 0, or None on error."""
    try:
        response = requests.get(url, headers=headers, timeout=timeout)
        response.raise_for_status()
        return int(keyword.lower() in response.text.lower())
    except RequestException as e:
        print(f"  Error: {e}")
        return None


def process_urls(input_csv, output_csv, keyword="fraud", delay=0.15):
    headers = {
        # REPLACE with your actual name and email
        "User-Agent": "Andrew Zilles andrew.zilles@ou.edu",
        "Accept-Encoding": "gzip, deflate",
        "Host": "www.sec.gov",
    }

    # Read URLs from input file — keep both the event key and the URL
    with open(input_csv, "r", newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)  # uses the header row as keys
        rows = [(r["aaer_event_key"], r["most_recent_release_url"]) for r in reader]

    print(f"Loaded {len(rows)} URLs from {input_csv}")

    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["aaer_event_key", "url", "fraud_flag"])

        for i, (event_key, url) in enumerate(rows, 1):
            print(f"[{i}/{len(rows)}] {url}")
            flag = check_url_for_keyword(url, keyword, headers)
            writer.writerow([event_key, url, flag])
            f.flush()
            time.sleep(delay)

    print(f"\nDone! Results saved to {output_csv}")


if __name__ == "__main__":
    base = r"C:/Users/andre/OneDrive - University of Oklahoma/3. Research/Tax AAER Replication/Tax AAER Replication"
    process_urls(
        input_csv=f"{base}/Fraud AAERs.csv",
        output_csv=f"{base}/results2.csv",
        keyword="fraud",
    )
```
